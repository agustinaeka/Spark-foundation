```python
data['Ship Mode'].value_counts()
```

```python
data.columns
```

```python
columns = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
for col in columns:
    plt.figure(figsize=(9,6))
    plt.pie(data[col].value_counts(), labels = data[col].value_counts().index, autopct='%.0f%%')
    plt.title(col)
    plt.show()
    
```

```python
byState = data.groupby('State')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
byState.plot.bar(figsize=(20,10))
```

 - California and New York have the highest sales and Profit

```python
byCat = data.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
byCat.plot.bar(figsize=(20,10))
```

- Although Copiers is less in sales compare to Phones but it has the highest profit

### Insight:


##### 1. The Profit generated by office supplies category is more, but the sales for office suppliers are less. So, we have to work on increasing     the sales for office supplies category.
##### 2. Work more on California and New York as they are the highest in sales
